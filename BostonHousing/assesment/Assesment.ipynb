```python
Excption = Excp_Data.unstack().reset_index()
Excption.columns = ['COMMENT', 'OUTPUT', 'Product']
```

```python
final = pd.merge(Pred_Data, Excption, how = 'left')
final['COMMENT'].fillna('SubProduct',inplace=True)
final
```

```python
final.sort_values(['COMMENT', 'Account.Number'],inplace=True)
final.reset_index(drop=True,inplace=True)
final['OUTPUT'] = True
```

```python
final
```

```python
final.loc[(final['Account.Number'].isin(['A4', 'A5']) & final['Product'].isin(['SP4'])), 'OUTPUT'] = 'NO'
final.loc[(final['Account.Number'].isin(['A3', 'A6']) & final['Product'].isin(['SP2'])), 'OUTPUT'] = 'NO'
```
